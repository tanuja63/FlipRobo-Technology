```python
#import all necesary libraries
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#load webdriver and get the website
driver=webdriver.Chrome("chromedriver.exe")
url="https://flight.yatra.com/air-search-ui/seodom/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=DEL&destinationCountry=IN&flexi=0&flight_depart_date=23%2F11%2F2021&hb=&noOfSegments=1&origin=BLR&originCountry=IN&type=O&unique=1634206221500&viewName=normal"
driver.get(url)
```

```python
airline=[]
source=[]
destination=[]
price=[]
stops=[]
route=[]
Dep_time=[]
arr_time=[]
date=[]
duration=[]
add_info=[]


try:
    for i in driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span"):
    
                airline.append(i.text)
except NoSuchElementException:
    airline.append("--")
except StaleElementReferenceException:
    airline.append("--")
except:
    airline.append("--")
try:
    

    for j in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
                Dep_time.append(j.text)
except NoSuchElementException:
    Dep_time.append("--")
except StaleElementReferenceException:
    Dep_time.append("--")
except:
    Dep_time.append("--")
try:
    
    
    
    for k in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]"):
    
                arr_time.append(k.text)
except NoSuchElementException:
    arr_time.append("--")
except StaleElementReferenceException:
    arr_time.append("--")
except:
    arr_time.append("--")
try:
    
    for l in driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/p"):
                duration.append(l.text)
except NoSuchElementException:
    duration.append("--")
except StaleElementReferenceException:
    duration.append("--")
except:
    duration.append("--")
try:
    
    
    
    for m in driver.find_elements_by_xpath("//div[@class='pr i-b mr-2']/div[2]"):
                price.append(m.text)
except NoSuchElementException:
    price.append("--")
except StaleElementReferenceException:
    price.append("--")
except:
    price.append("--")

try:
    
    

    for n in driver.find_elements_by_xpath("//div/div[1]/div[2]/div/div/div/div/div[1]/div/div[1]/p"):
    
                source.append(n.text)
except NoSuchElementException:
    source.append("--")
except StaleElementReferenceException:
    source.append("--")
except:
    source.append("--")
    
try:
    
    for o in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
                destination.append(o.text)
except NoSuchElementException:
    destination.append("--")
except StaleElementReferenceException:
    destination.append("--")
except:
    destination.append("--")
try:
    
    
    for p in driver.find_elements_by_xpath("//span[@class='cursor-default']|//span[@class='dotted-borderbtm']"):
                stops.append(p.text)
except NoSuchElementException:
    stops.append("--")
except StaleElementReferenceException:
     stops.append("--")
except:
     stops.append("--")
try:
    
    for q in driver.find_elements_by_xpath("//span[@class='text']"):
        add_info.append(q.text)
except NoSuchElementException:
     add_info.append("--")

except StaleElementReferenceException:
    add_info.append("--")
except:
    add_info.append("--")
    
for r in driver.find_elements_by_xpath("/html/body/section[2]/section/section[2]/section[1]/div[2]/div[2]/div/div/div/div[2]/div[2]/div[1]/div[2]/div/div/div/div[2]/div[1]/div[3]"):
                                        
                date.append(r.text)
    
```
