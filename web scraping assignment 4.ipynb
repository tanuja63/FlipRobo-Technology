# Question 1
. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#get the url
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/"
```

```python
driver.get(url)
```

```python
#let's scrape all data
rank=[]
name=[]
Artist=[]
date=[]
views=[]
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]"):
    rank.append(i.text)
for j in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]"):
    name.append(j.text)
for k in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]"):
    Artist.append(k.text)
for l in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]"):
    views.append(l.text)
for m in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]"):
    date.append(m.text)
```

```python
#create dataframe
Most_viewed_videos=pd.DataFrame({})
```

```python
Most_viewed_videos['Rank']=rank[:30]
Most_viewed_videos['Name']=name[:30]
Most_viewed_videos['Artist']=Artist[:30]
Most_viewed_videos['Uploaded date']=date[:30]
Most_viewed_videos['Views(in billions)']=views[:30]
Most_viewed_videos
```

```python
#scrap dataset name,data type,task,attribute type,no.of instance,no of attribute,year
name=[]
type1=[]
task=[]
attribute=[]
instance=[]
no_att=[]
year=[]
for i in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[1]//table//tbody//tr//td[2]//p//b//a"):
    try:
        
        name.append(i.text)
    except NoSuchElementException:
        name.append("--")
for j in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[2]/p"):
    try:
        type1.append(j.text)
    except NoSuchElementException:
        type1.append("--")

for k in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[3]/p"):
    try:
        task.append(k.text)
    except NoSuchElementException:
        task.append("--")
for l in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[4]/p"):
    try:
        attribute.append(l.text)
    except NoSuchElementException:
        attribute.append("--")
for m in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[5]/p"):
    try:
         instance.append(m.text)
    except NoSuchElementException:
        instance.append("--")
for n in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[6]/p"):
    try:
        no_att.append(n.text)
    except NoSuchElementException:
        no_att.append("--")
for o in driver.find_elements_by_xpath("//html//body//table[2]//tbody//tr//td[2]//table[2]//tbody//tr//td[7]/p"):
    try:
        year.append(o.text)
    except NoSuchElementException:
        year.append("--")
```

```python
t=[]
for i in range(0,len(type1),2):
    t.append(type1[i])
t
```

```python
#create dataframe and filling data into that dataframe
Datasets=pd.DataFrame({})
Datasets["Dataset Name"]=name
Datasets["Data Types"]=t[1:]
Datasets["Default Task"]=task[1:]
Datasets["Attributes Type"]=attributes[1:]
Datasets["Instances"]=instance[1:]
Datasets["Number of Attributes"]=no_att[1:]
Datasets["Year"]=year[1:]
Datasets
```
