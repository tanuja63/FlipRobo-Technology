```python
#import all necesary libraries
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#load webdriver and get the website
driver=webdriver.Chrome("chromedriver.exe")
url="https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001"
driver.get(url)
```

```python
#load all page urls
page_urls=[]

for i in range(1,150):

    t="https://www.cars24.com/buy-used-car?sort=P&page="
    t1="&storeCityId=2&pinId=110001"

    page_urls.append(''.join([t,str(i),t1]))
    
```

```python
urls=[]
for j in page_urls:
    driver.get(j)

    for i in driver.find_elements_by_xpath("//a[@class='_9Ue0B']"):
         urls.append(i.get_attribute('href'))
       
```

```python
#scrape model_name,brand_name,location,price,transmiision,num_of_ownres,fuel from website
model_name=[]
brand=[]
location=[]
price=[]
no_of_owners=[]
fuel=[]
driven_kmtrs=[]
for j in urls[:110]:
    driver.get(j)
    try:
        b=driver.find_element_by_xpath("//h2[@class='_2geSF']")
        brand.append(b.text)
    except NoSuchElementException:
        brand.append("--")
    except StaleElementReferenceException:
        brand.append("--")
    except:
        brand.append("--")
    try:
        m=driver.find_element_by_xpath("//span[@class='_18vo2']")
        model_name.append(m.text)
    except NoSuchElementException:
        model_name.append("--")
    except StaleElementReferenceException:
        model_name.append("--")
    except:
        model_name.append("--")
    try:
        d=driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[1]")
        
        driven_kmtrs.append(d.text)
    except NoSuchElementException:
        driven_kmtrs.append("--")
    except StaleElementReferenceException:
        driven_kmtrs.append("--")
    except:
        driven_kmtrs.append("--")
    try:
        f=driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[3]")
        
        fuel.append(f.text)
    except NoSuchElementException:
        fuel.append("--")
    except StaleElementReferenceException:
        fuel.append("--")
    except:
        fuel.append("--")
    try:
        o=driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[5]")
        
        
        no_of_owners.append(o.text)
    except NoSuchElementException:
        no_of_owners.append("--")
    except StaleElementReferenceException:
        no_of_owners.append("--")
    except:
        no_of_owners.append("--")
    try:
        p=driver.find_element_by_xpath("//strong[@class='_2yYvS']")
        
        
        
        price.append(p.text)
    except NoSuchElementException:
        price.append("--")
    except StaleElementReferenceException:
        price.append("--")
    except:
        price.append("--")
    try:
        l=driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[7]")
       
        
        
        
        location.append(l.text)
    except NoSuchElementException:
        location.append("--")
    except StaleElementReferenceException:
        location.append("--")
    except:
        location.append("--")
    
print(len(brand))
print(len(model_name))
print(len(location))
print(len(price))
print(len(no_of_owners))
print(len(fuel))
print(len(driven_kmtrs))
```
